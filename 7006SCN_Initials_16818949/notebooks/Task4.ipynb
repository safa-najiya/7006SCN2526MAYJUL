```python
from pyspark.sql import SparkSession
from pyspark import StorageLevel
```

# Creating Spark Session

```python
spark = SparkSession.builder \
    .appName("AmazonReviews_Task4") \
    .master("local[*]") \
    .config("spark.executor.memory", "4g") \
    .config("spark.driver.memory", "4g") \
    .config("spark.sql.shuffle.partitions", "8") \
    .getOrCreate()

print("Spark Version:", spark.version)
```

# Data Partitioning, Caching and Persisting

```python
df = spark.read.parquet("amazon_reviews_clean_11M")

print("Rows:", df.count())
print("Partitions:", df.rdd.getNumPartitions())
```

```python
print("Before:", df.rdd.getNumPartitions())

df = df.coalesce(8)

print("After:", df.rdd.getNumPartitions())
```

```python
sample_df = df.limit(500000)
```

```python
sample_df.cache()

sample_df.count()

print("Sample Dataset Cached Successfully")
```

```python
print(sample_df.storageLevel)
```

```python
from pyspark import StorageLevel

sample_df.persist(StorageLevel.MEMORY_AND_DISK)

sample_df.count()

print(sample_df.storageLevel)
```
