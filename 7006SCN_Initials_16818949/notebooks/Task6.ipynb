```python
from pyspark.sql import SparkSession
```

# Creating Spark Session

```python
spark = SparkSession.builder \
    .appName("AmazonReviews_Task6") \
    .master("local[*]") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

print("Spark Version:", spark.version)
```

# Converting Parquet files to CSV and Dowloading

```python
df = spark.read.parquet("amazon_reviews/amazon_reviews_clean_11M")
```

```python
print("Rows:", df.count())

df.printSchema()

df.show(5, truncate=False)
```

```python
df_csv = df.drop("images")
```

```python
from pyspark.sql import functions as F

df_tableau = df_csv.orderBy(F.rand(seed=42)).limit(250000)

df_tableau.coalesce(1) \
    .write \
    .mode("overwrite") \
    .option("header", True) \
    .csv("tableau_dataset")
```
